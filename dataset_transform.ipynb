## Dependencies

```python
import pandas as pd
import os
```

## Tranform data

```python
def transform_data(eng_path,ger_path):

    with open(eng_path, "r", encoding="utf-8") as eng_file:
        english_file = eng_file.readlines()

    with open(ger_path, "r", encoding="utf-8") as ger_file:
        german_file = ger_file.readlines()

    data = {
        "English": [line.strip() for line in english_file],
        "German": [line.strip() for line in german_file],
    }
    return pd.DataFrame(data)

def save_file(df,filename,output_path = "./data"):
    filepath = os.path.join(output_path,filename)
    return df.to_csv(path_or_buf=filepath,index=False,encoding="utf-8")


```

### Train data

```python
train_data = transform_data(eng_path="./data/train.en",ger_path="./data/train.de")
save_file(df=train_data,filename="train_data.csv")
```

### Validation data

```python
val_data = transform_data(eng_path="./data/val.en",ger_path="./data/val.de")
save_file(df=val_data,filename="val_data.csv")
```

### Test data

```python
test_data = transform_data(eng_path="./data/test.en",ger_path="./data/test.de")

with open("./data/test.fr","r",encoding="utf-8") as fr_file:
    french_file = fr_file.readlines()

test_data['French'] = [line.strip() for line in french_file]
save_file(df=test_data,filename="test_data.csv")
```
